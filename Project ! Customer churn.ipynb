### 1. Data Manipulation:

● Extract all those customers whose tenure is greater than 70 months or

```
customer_5 = df.iloc[:, 4]
print(customer_5.head())
```

```
senior_male_electronic = df[(df["gender"] == "Male") &
                            (df["SeniorCitizen"] == 1) &
                            (df["PaymentMethod"] == "Electronic check")]
print(senior_male_electronic.head())
senior_male_electronic.to_csv("senior_male_electronic.csv", index=False)
```

```
customer_total_tenure = df[(df['tenure'] > 70) | (df['MonthlyCharges'] > 100)]
customer_total_tenure.to_csv("customer_total_tenure.csv", index=False)
print(customer_total_tenure.head())
```

Extract all the customers whose contract is of two years, payment method
is mailed check and the value of churn is ‘Yes’ and store the result in
‘two_mail_yes’


```
two_mail_yes = df[
    (df["Contract"] == "Two year") &
    (df["PaymentMethod"] == "Mailed check") &
    (df["Churn"] == "Yes")
]
print(two_mail_yes)
two_mail_yes.to_csv("two_mail_yes.csv", index=False)
```

### 2. Data Visualization:


### ● Build a bar-plot for the ’InternetService’ column:

### a. Set x-axis label to ‘Categories of Internet Service

```
plt.figure(figsize=(8, 5))
df['InternetService'].value_counts().plot(kind='bar', color=['blue', 'green', 'red'])

# Set labels and title
plt.xlabel("Categories of Internet Service")
plt.ylabel("Count")
plt.title("Distribution of Internet Service Categories")

# Show the plot
plt.show()
```

Set y-axis label to ‘Count of Categories

```
plt.figure(figsize=(8, 5))
df['InternetService'].value_counts().plot(kind='bar', color=['blue', 'orange', 'green'])

# Set labels
plt.xlabel('Internet Service Type')
plt.ylabel('Count of Categories')
plt.title('Distribution of Internet Service Types')

# Show the plot
plt.show()
```

Set the title of plot to be ‘Distribution of Internet Service

```
plt.figure(figsize=(8, 5))
df['InternetService'].value_counts().plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.title('Distribution of Internet Service', fontsize=14)
plt.xlabel('Internet Service Type', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show plot
plt.show()
```

Build a scatter-plot between ‘MonthlyCharges’ and ‘tenure’. Map
‘MonthlyCharges’ to the y-axis and ‘tenure’ to the ‘x-axis’:
a. Assign the points a color of ‘brown

```
plt.figure(figsize=(8, 6))
plt.scatter(df['tenure'], df['MonthlyCharges'], color='brown', alpha=0.5)
plt.xlabel('Tenure (Months)')
plt.ylabel('Monthly Charges ($)')
plt.title('Scatter Plot of Monthly Charges vs Tenure')
plt.grid(True)
plt.show()
```

Set the x-axis label to ‘Tenure of customer’

```
plt.figure(figsize=(8,6))
plt.scatter(df['tenure'], df['MonthlyCharges'], alpha=0.5, color='blue')
plt.xlabel('Tenure of customer')
plt.ylabel('Monthly Charges')
plt.title('Scatter Plot of Monthly Charges vs Tenure')

# Show the plot
plt.show()
```

Set the y-axis label to ‘Monthly Charges of customer’

```
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['tenure'], y=df['MonthlyCharges'], alpha=0.5)
plt.xlabel('Tenure of Customer (Months)')
plt.ylabel('Monthly Charges of Customer')
plt.title('Scatter Plot of Monthly Charges vs Tenure')

# Show plot
plt.show()
```

Set the title to ‘Tenure vs Monthly Charges’

```
plt.figure(figsize=(8, 6))
plt.scatter(df['tenure'], df['MonthlyCharges'], alpha=0.5, color='blue')
plt.xlabel('Tenure (Months)')
plt.ylabel('Monthly Charges ($)')
plt.title('Tenure vs Monthly Charges')
plt.grid(True)

# Show plot
plt.show()
```

Build a box-plot between ‘tenure’ & ‘Contract’. Map ‘tenure’ on the y-axis &

```
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['tenure'], y=df['MonthlyCharges'], alpha=0.5)
plt.xlabel('Tenure')
plt.ylabel('Monthly Charges')
plt.title('Scatter Plot of Monthly Charges vs. Tenure')
plt.show()


# Box plot between tenure (y-axis) and Contract
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Contract'], y=df['tenure'])
plt.xlabel('Contract Type')
plt.ylabel('Tenure')
plt.title('Box Plot of Tenure by Contract Type')
plt.xticks(rotation=45)
plt.show()
```

'Contract' on the X-Axis

```
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="tenure", y="MonthlyCharges", hue="Contract", alpha=0.7)
plt.xlabel("Tenure (Months)")
plt.ylabel("Monthly Charges ($)")
plt.title("Scatter Plot of Monthly Charges vs. Tenure, Colored by Contract Type")
plt.legend(title="Contract Type")
plt.show()
```
